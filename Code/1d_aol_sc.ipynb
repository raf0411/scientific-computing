```python
import matplotlib.pyplot as plt

year = [1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999]

temp = [14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438]

years = ['1981', '1983', '1985', '1987', '1989', '1991', '1993', '1995', '1997', '1999']

plt.plot(year, temp)

plt.ylim(14.0 , 14.5)
plt.xlim(1981 , 1999)

plt.title("Relationship between Temperature and Year")
plt.ylabel('Temperature (°C)')
plt.xlabel('Year')

plt.xticks(year, years)
plt.rcParams.update({'font.size': 8, 'font.serif':'Arial'})
```
